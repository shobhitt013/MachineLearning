```python
import numpy as np
import pandas as pd
df = pd.read_csv('/content/customer.csv')
df.sample(5)
```

```python
df = df.iloc[:,2:]
df.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
X_train = df.copy()
X_train
```

```python
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
# Only fit on the first two columns (review and education) to match the categories provided
oe.fit(X_train.iloc[:, :2])
```

```python
X_train.iloc[:, :2] = oe.transform(X_train.iloc[:, :2])
X_train
```

```python
oe.categories_
```

```python
X_train
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_train = X_train['purchased']
le.fit(y_train)
```

```python
le.classes_
```

```python
y_train = X_train['purchased']
y_train = le.transform(y_train)
y_train
```
